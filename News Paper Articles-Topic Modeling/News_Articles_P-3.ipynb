```python
df=site_article.text
df
```

```python
# Preprocessing train dataset
import re
import nltk
import string
from string import punctuation
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
lemma=WordNetLemmatizer()
```

```python
final_df=[]
for i in range(0,len(df)):
    review=re.sub('[^a-zA-Z]',' ', df[i])
    review=review.lower()
    review=review.split()
    review=[lemma.lemmatize(word) for word in review if not word in string.punctuation if not word in stopwords.words('english')]
    review=' '.join(review)
    final_df.append(review)
```

```python
import gensim
from nltk import word_tokenize, sent_tokenize
```

```python
for i in range (0,len(final_df)):
    final_df[i]=word_tokenize(final_df[i])
final_df
```

```python
dictionary=gensim.corpora.Dictionary(final_df)
print(dictionary)
```

```python
dictionary.filter_extremes(no_below=5,no_above=0.2,keep_n=100000)
```

```python
print(dictionary)
```

```python
bow_corpus=[dictionary.doc2bow(doc) for doc in final_df]
```

```python
bow_corpus
```

```python
lda_model=gensim.models.LdaMulticore(bow_corpus,num_topics=30,id2word=dictionary,passes=10,workers=2)
```

```python
for idx, topic in lda_model.print_topics():
    print("Topic: {} \nWords: {}".format (idx,topic))
    print("")
```
